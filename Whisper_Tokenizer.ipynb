```python
prompt = input("Введите промпт: ")
```

```python
from whisper.tokenizer import get_tokenizer

tokenizer = get_tokenizer(multilingual=True)
tokens = tokenizer.encode(prompt)

limit = 224
fits = len(tokens) <= limit

print(f"Токены: {len(tokens)}/{limit} ({len(tokens) / limit * 100:.2f}%)")

if not fits:
    kept_tokens = tokens[-limit:]
    kept_text = tokenizer.decode(kept_tokens)

    print("\nВ лимит попадёт только эта часть:")
    print(kept_text)

print("\n======================================================================")

for i, t in enumerate(tokens):
    print(i, t, repr(tokenizer.decode([t])))
```
